```python
genders=df['Gender']
ages=df['Age']
genders_counts = genders.value_counts()
plt.figure(figsize=(8,6))
genders_counts.plot(kind='bar',color=['blue','pink','purple'])
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender distribution in the population')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.hist(ages,bins=30,color='orange',edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('age distribution in the population')
plt.show()
```
